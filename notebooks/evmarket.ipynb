**User**: Load './data/Electric_Vehicle_Population_Data.csv'

```python
import pandas as pd

data = pd.read_csv('../data/Electric_Vehicle_Population_Data.csv')
data.head()
```

**User**: List the top seven models, along with others combining all other models.

```python
top_models = data['Model'].value_counts().head(7)
other_models_count = data['Model'].value_counts().sum() - top_models.sum()

top_models['Other'] = other_models_count

top_models
```

```python
import matplotlib.pyplot as plt

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_models, labels=top_models.index, autopct='%1.1f%%')
plt.title('Electric Vehicle Models')

# Display the chart
plt.show()
```
