**User**: 渋谷駅周辺のスタバを表示

```python
import requests
import folium

# 渋谷駅の座標を取得
response = requests.get("https://nominatim.openstreetmap.org/search?q=渋谷駅&format=json")
data = response.json()
lat = float(data[0]["lat"])
lon = float(data[0]["lon"])

# Overpass APIを使ってスターバックスのデータを取得
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = '[out:json];node(around:1000, {lat}, {lon})["amenity"="cafe"]["name"~"スターバックス"];out;'.format(
    lat=lat, lon=lon
)
response = requests.get(overpass_url, params={"data": overpass_query})
data = response.json()

# 地図を作成してスターバックスの位置を表示
m = folium.Map(location=[lat, lon], zoom_start=15)
for node in data["elements"]:
    name = node["tags"]["name"]
    lat = node["lat"]
    lon = node["lon"]
    folium.Marker([lat, lon], popup=name).add_to(m)

m
```

**User**: 渋谷駅周辺のスターバックスとドトールの位置を表示する

```python
import requests
import folium

# 渋谷駅の座標を取得
response = requests.get("https://nominatim.openstreetmap.org/search?q=渋谷駅&format=json")
data = response.json()
lat = float(data[0]["lat"])
lon = float(data[0]["lon"])

# Overpass APIを使ってスターバックスのデータを取得
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = '[out:json];node(around:1000, {lat}, {lon})["amenity"="cafe"]["name"~"スターバックス"];out;'.format(
    lat=lat, lon=lon
)
response = requests.get(overpass_url, params={"data": overpass_query})
starbucks_data = response.json()

# Overpass APIを使ってドトールのデータを取得
overpass_query = '[out:json];node(around:1000, {lat}, {lon})["amenity"="cafe"]["name"~"ドトール"];out;'.format(
    lat=lat, lon=lon
)
response = requests.get(overpass_url, params={"data": overpass_query})
dotour_data = response.json()

# 地図を作成してスターバックスとドトールの位置を表示
m = folium.Map(location=[lat, lon], zoom_start=15)

# スターバックスの位置を表示
for node in starbucks_data["elements"]:
    name = node["tags"]["name"]
    lat = node["lat"]
    lon = node["lon"]
    folium.Marker([lat, lon], popup=name, icon=folium.Icon(color="green")).add_to(m)

# ドトールの位置を表示
for node in dotour_data["elements"]:
    name = node["tags"]["name"]
    lat = node["lat"]
    lon = node["lon"]
    folium.Marker([lat, lon], popup=name, icon=folium.Icon(color="blue")).add_to(m)

m
```
