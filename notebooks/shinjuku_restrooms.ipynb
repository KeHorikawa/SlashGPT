**User**: 新宿駅周辺の公衆トイレをOverpass APIで見つけて、foliumを使って地図上に表示して。

```python
import requests
import folium

# Overpass APIのエンドポイント
url = 'https://overpass-api.de/api/interpreter'

# Overpass QLクエリ
query = '''
[out:json];
node(around:1000, 35.6895, 139.7005)[amenity=toilets];
out;
'''

# Overpass APIにリクエストを送信
response = requests.get(url, params={'data': query})

# レスポンスのJSONデータを取得
data = response.json()

# 地図を作成
m = folium.Map(location=[35.6895, 139.7005], zoom_start=15)

# 公衆トイレの位置を地図上に表示
for toilet in data['elements']:
    lat = toilet['lat']
    lon = toilet['lon']
    folium.Marker([lat, lon]).add_to(m)

# 地図を表示
m
```
